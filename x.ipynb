```python
import pandas as pd

df = pd.read_csv("cardio_base.csv")
```

```python
df['age_years'] = df['age'] // 365
```

```python
avg_weight_by_age = df.groupby('age_years')['weight'].mean()
```

```python
max_weight_age = avg_weight_by_age.idxmax()
min_weight_age = avg_weight_by_age.idxmin()
```

```python
max_avg_weight = avg_weight_by_age[max_weight_age]
min_avg_weight = avg_weight_by_age[min_weight_age]

percentage_difference = ((max_avg_weight - min_avg_weight) / min_avg_weight) * 100
```

```python
print(f"The age group with the highest average weight is {max_weight_age} years.")
print(f"The age group with the lowest average weight is {min_weight_age} years.")
print(f"The weight difference is approximately {percentage_difference:.2f}%.")
```

```python
over_50 = df[df['age_years'] > 50]
under_50 = df[df['age_years'] <= 50]
```

```python
avg_cholesterol_over_50 = over_50['cholesterol'].mean()
avg_cholesterol_under_50 = under_50['cholesterol'].mean()
```

```python
print(f"The average cholesterol level for people over 50 is {avg_cholesterol_over_50:.2f}.")
print(f"The average cholesterol level for people under 50 is {avg_cholesterol_under_50:.2f}.")

if avg_cholesterol_over_50 > avg_cholesterol_under_50:
    print("People over 50 have higher cholesterol levels.")
else:
    print("There is no significant difference in cholesterol levels between the two age groups.")
```

```python

avg_cholesterol_over_50 = df[df['age_years'] > 50]['cholesterol'].mean()
avg_cholesterol_under_50 = df[df['age_years'] <= 50]['cholesterol'].mean()

percentage_difference = ((avg_cholesterol_over_50 - avg_cholesterol_under_50) / avg_cholesterol_under_50) * 100

print(f"The average cholesterol level for people over 50 is {avg_cholesterol_over_50:.2f}.")
print(f"The average cholesterol level for people under 50 is {avg_cholesterol_under_50:.2f}.")

if percentage_difference > 0:
    print(f"People over 50 have higher cholesterol levels.")
    print(f"Yes, their cholesterol level is {abs(percentage_difference):.0f}% higher on average.")
else:
    print("No, it's about the same.")
```

```python
smoking_column = 'smoke'

smoking_ratio_men = df[df['gender'] == 'male'][smoking_column].mean()
smoking_ratio_women = df[df['gender'] == 'female'][smoking_column].mean()

smoking_ratio = smoking_ratio_men / smoking_ratio_women

if smoking_ratio > 1:
    print(f"Yes, men are {smoking_ratio:.0f}x more likely to be smokers.")
elif smoking_ratio < 1:
    print(f"No, women are {1/smoking_ratio:.0f}x more likely to be smokers.")
else:
    print("Similar portions of men and women are smokers.")
```

```python
import pandas as pd

df_alco = pd.read_csv("cardio_alco.csv")

df['id'] = df['id'].astype(str)
df_alco['id'] = df_alco['id'].astype(str)

merged_df = pd.merge(df, df_alco, on='id', how='inner')

over_50_df = merged_df[merged_df['age_years'] > 50]

percentage_alcohol_consumers_over_50 = (over_50_df['alco'] == 1).mean() * 100

print(f"The percentage of the population over 50 years old that consumes alcohol is approximately {percentage_alcohol_consumers_over_50:.0f}%.")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

blood_pressure_men = df[df['gender'] == 'male']['cholesterol']
blood_pressure_women = df[df['gender'] == 'female']['smoke']

t_stat, p_value = ttest_ind(blood_pressure_men, blood_pressure_women, equal_var=False)

if p_value < 0.05:
    print("With 95% confidence, men have higher blood pressure than women.")
else:
    print("There is not enough evidence to conclude that men have higher blood pressure than women.")

```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

df_covid19 = pd.read_csv("covid_data.csv")

italy_data = df_covid19[(df_covid19['location'] == 'Italy') & (df_covid19['date'] >= '2020-02-28') & (df_covid19['date'] <= '2020-03-20')]

def exponential_growth(x, A, B):
    return A * np.exp(B * x)

italy_data['days_since_start'] = (pd.to_datetime(italy_data['date']) - pd.to_datetime('2020-02-28')).dt.days

params, covariance = curve_fit(exponential_growth, italy_data['days_since_start'], italy_data['new_cases'])

italy_data['predicted_cumulative_cases'] = italy_data['days_since_start'].apply(lambda x: exponential_growth(x, *params)).cumsum()

predicted_total_cases_on_2020_03_20 = italy_data.loc[italy_data['date'] == '2020-03-20', 'predicted_cumulative_cases'].values[0]

real_total_cases_on_2020_03_20 = italy_data.loc[italy_data['date'] == '2020-03-20', 'new_cases'].cumsum().values[0]

difference = abs(predicted_total_cases_on_2020_03_20 - real_total_cases_on_2020_03_20)
```

```python
print(f"The difference between the predicted cumulative cases and the total number of real cases on 2020-03-20 is approximately {difference:.0f}.")
```

```python
import pandas as pd
df_covid19 = pd.read_csv("covid_data.csv")

df_covid19['death_rate_per_million'] = (df_covid19['new_deaths'] / df_covid19['population']) * 1_000_000

total_death_rate = df_covid19.groupby('location')['death_rate_per_million'].sum()

sorted_countries = total_death_rate.sort_values(ascending=False)

third_highest_country = sorted_countries.index[2]

print(f"The country with the 3rd highest death rate is: {third_highest_country}.")
```

```python
import pandas as pd
from sklearn.metrics import f1_score

df_covid19 = pd.read_csv("covid_data.csv")

df_covid19['positive_condition'] = (df_covid19['aged_65_older_percent'] > 20) & (df_covid19['new_deaths'] > 50)
df_covid19['negative_condition'] = ~df_covid19['positive_condition']

df_covid19['missing_info'] = df_covid19[['aged_65_older_percent', 'new_deaths', 'population']].isnull().any(axis=1)

df_covid19 = df_covid19[~df_covid19['missing_info']]

true_labels = df_covid19['positive_condition']
predicted_labels = df_covid19['negative_condition']  # Assuming we predict all countries as negative for simplicity

f1 = f1_score(true_labels, predicted_labels)

print(f"The F1 score for the given statement is: {f1:.2f}")
```
